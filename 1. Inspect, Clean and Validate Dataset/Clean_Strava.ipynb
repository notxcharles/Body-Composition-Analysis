```python
import pandas as pd
```

```python
strava_df = pd.read_csv("../DataInput/Strava/activities.csv")
strava_df.head()
```

subset of the columns

```python
strava_df = strava_df[["Activity Date", "Activity Name", "Activity Type", "Distance", "Calories"]]
```

```python
strava_df
```

Here, calories marked as NaN are useful- they signify that an activity has occurred in that day, but the data is missing.

Lets adjust activity date column into dd/mm/yyyy

```python
activity_date = strava_df["Activity Date"].apply(lambda s: " ".join(s.split(" ")[:3]))
activity_date = activity_date.apply(lambda s: s[:-1])
activity_date = pd.to_datetime(activity_date)
activity_date.head()
```

```python
strava_df["Activity Date"] = activity_date
strava_df
```

Dateset should be cleaned, save to file

```python
strava_df.to_csv("./CleanedData/Cleaned_Strava.csv")
```
