```python
import pandas as pd
```

```python
mfp_df = pd.read_csv("../DataInput/MyFitnessPal/Nutrition-Summary-2020-05-19-to-2025-08-15.csv")
```

I can't remember using MFP for the first 2 rows of data, so I'm going to remove them

```python
mfp_df = mfp_df[2:]
mfp_df
```

We only care about the Date, Calories, Fat, Carbohydrates and Protein columns.
- A lot of foods that I have logged have unclear amounts of micronutrients
- The specific meal that the calories was consumed in is inaccurate in a lot of cases and doesn't represent one meal

```python
mfp_df = mfp_df[["Date", "Calories", "Fat (g)", "Carbohydrates (g)", "Protein (g)"]]
mfp_df
```

There are 6 rows that appear to have missing data. I'm choosing to keep these rows in the dataset because it would represent having 0 calories to eat that day

```python
na_rows = mfp_df[mfp_df.isna().any(axis=1)]
na_rows
```

```python
mfp_df = mfp_df.fillna(0)
mfp_df[mfp_df.isna().any(axis=1)]
```
